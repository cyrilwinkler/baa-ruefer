```python
# count number of zeros in a list
def count_zeros(list):
    count = 0
    for i in list:
        if i == 0:
            count += 1
    return count

print(f'rouge_1 : {round(count_zeros(rouge_1)/len(rouge_1), 2)}%')
print(f'rouge_2 : {round(count_zeros(rouge_2)/len(rouge_2), 2)}%')
print(f'rouge_l : {round(count_zeros(rouge_l)/len(rouge_l), 2)}%')
```

```python
import json
from rouge import Rouge

with open(f'final_output_emotional.json', 'r') as f:
    output_data = json.load(f)

with open('llama_output_old.json', 'r') as f:
    input_json = json.load(f)

team_ids = [487, 489]
rouge_scores_1 = []
rouge_scores_2 = []
rouge_scores_l = []

def calculate_rouge_scores(hypothesis, reference):
    rouge = Rouge()
    scores = rouge.get_scores(hypothesis, reference)
    rouge_scores_1.append(scores[0]['rouge-1']['f'])
    rouge_scores_2.append(scores[0]['rouge-2']['f'])
    rouge_scores_l.append(scores[0]['rouge-l']['f'])
```

```python
for team_id in team_ids:
    team_id = str(team_id)
    model_outputs = output_data[str(team_id)]['information']
    if type(model_outputs) == dict:
        for output in model_outputs.values():
            calculate_rouge_scores(output, input_json[str(team_id)]['information'])
    else:
        rouge_scores_1.append(0)
        rouge_scores_2.append(0)
        rouge_scores_l.append(0)

    for statistic in input_json[team_id]['statistics']:
        try:
            team_statistics_output = output_data[str(team_id)]['statistics'][statistic]
            if type(team_statistics_output) == dict:
                for output in team_statistics_output.values():
                    calculate_rouge_scores(output, input_json[str(team_id)]['statistics'][statistic])
            else:
                rouge_scores_1.append(0)
                rouge_scores_2.append(0)
                rouge_scores_l.append(0)
        except:
            rouge_scores_1.append(0)
            rouge_scores_2.append(0)
            rouge_scores_l.append(0)

    for player_id in input_json[team_id]['players']:
        try:
            player_information_output = output_data[str(team_id)]['players'][player_id]['information']
            if type(player_information_output) == dict:
                for output in player_information_output.values():
                    calculate_rouge_scores(output, input_json[str(team_id)]['players'][player_id]['information'])
            else:
                rouge_scores_1.append(0)
                rouge_scores_2.append(0)
                rouge_scores_l.append(0)
        except:
            rouge_scores_1.append(0)
            rouge_scores_2.append(0)
            rouge_scores_l.append(0)

        for statistic in output_data[str(team_id)]['players'][player_id]['statistics']:
            try:
                player_statistics_output = output_data[str(team_id)]['players'][player_id]['statistics'][statistic]
                if type(player_statistics_output) == dict:
                    for output in player_statistics_output.values():
                        calculate_rouge_scores(output, input_json[str(team_id)]['players'][player_id]['statistics'][statistic])
                else:
                    rouge_scores_1.append(0)
                    rouge_scores_2.append(0)
                    rouge_scores_l.append(0)
            except:
                rouge_scores_1.append(0)
                rouge_scores_2.append(0)
                rouge_scores_l.append(0)

    coach_output = output_data[str(team_id)]['coach']
    if type(coach_output) == dict:
        try:
            for output in coach_output.values():
                calculate_rouge_scores(output, input_json[str(team_id)]['coach']['information'])
        except:
            rouge_scores_1.append(0)
            rouge_scores_2.append(0)
            rouge_scores_l.append(0)
    else:
        rouge_scores_1.append(0)
        rouge_scores_2.append(0)
        rouge_scores_l.append(0)

print(f"ROUGE-1: {sum(rouge_scores_1)/len(rouge_scores_1)}")
print(f"ROUGE-2: {sum(rouge_scores_2)/len(rouge_scores_2)}")
print(f"ROUGE-L: {sum(rouge_scores_l)/len(rouge_scores_l)}")


```
